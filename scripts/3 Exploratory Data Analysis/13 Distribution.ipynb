### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# 1. Column names
print('Columns:')
list(df.columns)

# 2. Data Types
print('\nData types:')
print(df.dtypes)

# 3. Summary info
print('\nDataFrame info:')
df.info()

# 4. Summary Stats
print('\nNumeric summary:')
print(df.describe(include='all'))
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

# Identify missing values
missing_percent = (df.isna().sum() / len(df)) * 100
missing_percent = missing_percent.sort_values(ascending=False)
print('Missing percentages per column(before cleaning):')
#pd.set_option('display.max_rows', None)
print(missing_percent)
#pd.reset_option('display.max_rows')

# Drop non critical columns
cols_to_drop = missing_percent[(missing_percent>=60)].index
df = df.drop(columns=cols_to_drop)
print(f'\nDropped columns(non-critical, >=60% missing): {list(cols_to_drop)}')


# Handle remaining missing values
    # a. Numeric columns >>> fill with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

    # b. Categorical Columns >>> fill missing with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify missing values are handled
missing_percent_after = (df.isna().sum() / len(df)) * 100
missing_percent_after = missing_percent_after.sort_values(ascending=False)
print('\nMissing Percentages per column(after cleaning):')
#pd.set_option('display.max_rows', None)
print(missing_percent_after)
#pd.reset_option('display.max_rows')
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
for name in ['Employment', 'JobSat', 'YearsCodePro']:
    print(f'\nValue counts for column: {name}')
    print(df[name].value_counts())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here

#job_counts = df['JobSat'].value_counts().sort_index()
plt.figure(figsize=(8,6))
sns.kdeplot(df['JobSat'], fill=True, bw_adjust=0.5, color='skyblue')
plt.title('Job Satisfaction Distribution (KDE)')
plt.xlabel('Job Satisfaction')
plt.ylabel('Density')
plt.show()

print('The Job Satisfaction scales from 0-10. With a clear mean of 7, as it shows on the graph as extremely dense at Job Satisfaction of 7, then smoothes out.')
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# split multi-select entries into lists
df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].str.split(';')
df['LanguageWantToWorkWith'] = df['LanguageWantToWorkWith'].str.split(';')

# explode columns to get one language per row
have_worked = df.explode('LanguageHaveWorkedWith')['LanguageHaveWorkedWith'].str.strip()
want_work = df.explode('LanguageWantToWorkWith')['LanguageWantToWorkWith'].str.strip()

# get value counts
have_counts = have_worked.value_counts()
want_counts = want_work.value_counts()

# send lang_counts to a dataframe
lang_counts = pd.DataFrame({
    'HaveWorkedWith': have_counts,
    'WantToWorkWith': want_counts
}).sort_values(by='HaveWorkedWith', ascending=False)

# use the top 10 languages for plotting
top_n = 20
top_langs = lang_counts.head(top_n)

top_langs.plot(kind='bar', figsize=(12,6))
plt.title('Frequency of Programming Languages: Have/Want to Work With')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

#group and count by Country and RemoteWork type
remote_counts = df.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')
remote_pivot = remote_counts.pivot(index='Country', columns='RemoteWork', values='Count').fillna(0)

country_counts = df['Country'].value_counts()
top_countries = country_counts.head(20).index
df_top = df[df['Country'].isin(top_countries)]
remote_counts_top = df_top.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')
remote_pivot_top = remote_counts_top.pivot(index='Country', columns='RemoteWork', values='Count').fillna(0)

remote_pivot_top.plot(
    kind='bar',
    figsize=(14,7),
    stacked=False,
    colormap='tab20'
)
plt.title('Remote Work Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
# Ensure numeric
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 50
})
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
#df['YearsCodePro'].dtype

pearson_corr = df['JobSat'].corr(df['YearsCodePro'],method='pearson')
print(f'Pearson correlation: {pearson_corr:.3f}')
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
# Clean / explode data
df_ed = df.dropna(subset=['EdLevel', 'Employment']).copy()
df_ed['Employment'] = df_ed['Employment'].str.split(';')
df_ed = df_ed.explode('Employment')
df_ed['Employment'] = df_ed['Employment'].str.strip()

ed_emp_crosstab = pd.crosstab(
    df_ed['EdLevel'],
    df_ed['Employment'],
)
#print(ed_emp_crosstab)

ed_emp_crosstab.plot(
    kind='bar',
    stacked=True,
    figsize=(14,7),
    colormap='tab20'
)
plt.title('Employment Type Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
