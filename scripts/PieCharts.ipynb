```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_csv('survey_data.csv')

df.head()
```

```python
## Visualizing Data Composition with Pie Charts
# 1. Create a Pie Chart of the Top 5 Databases Respondents Want to Work With

df_1 = df[['DatabaseWantToWorkWith']].dropna()

df_1['DatabaseWantToWorkWith'] = df_1['DatabaseWantToWorkWith'].str.split(';')
df_1 = df_1.explode('DatabaseWantToWorkWith')

df_1_counts = df_1['DatabaseWantToWorkWith'].value_counts()
df_1_top5 = df_1_counts.head(5)

plt.figure(figsize=(8,8))
plt.pie(
    df_1_top5,
    labels=df_1_top5.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Top 5 Databases Resondents Want To Work With')
plt.show()
```

```python
# 2. Top 5 Developer Roles

df_1_2 = df[['DevType']].copy()
df_1_2 = df_1_2.dropna()

top5_devType = df_1_2['DevType'].value_counts().head(5)


plt.figure(figsize=(8,8))
plt.pie(
    top5_devType,
    labels=top5_devType.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Top 5 Developer Types')
plt.show()
```

```python
# 3. Pie Chart for the Operating Systems used by Respondents for Professional Use

df_1_3 = df[['OpSysProfessional use']].copy()
df_1_3 = df_1_3.dropna()

df_1_3['OpSysProfessional use'] = df_1_3['OpSysProfessional use'].str.split(';')
df_1_3 = df_1_3.explode('OpSysProfessional use')

top5_os = df_1_3['OpSysProfessional use'].value_counts().head(5)
print(top5_os)


plt.figure(figsize=(8,8))
plt.pie(
    top5_os,
    labels=top5_os.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Top 5 Professional Operating Systems')
plt.show()
```
